```python
from keras import models
from keras import layers
from keras.layers import Flatten, Dropout
from keras import optimizers, regularizers

AlexNet = models.Sequential()
AlexNet.add(layers.Conv2D(96,(11,11), strides = 4, padding = 'valid', input_shape =(227, 227, 3)))
AlexNet.add(layers.BatchNormalization())
AlexNet.add(layers.Activation('relu'))

AlexNet.add(layers.MaxPooling2D(pool_size = (3,3), strides = 2, padding = 'valid'))

AlexNet.add(layers.Conv2D(256,(5,5), strides = 1, padding = 'same' ))
AlexNet.add(layers.BatchNormalization())
AlexNet.add(layers.Activation('relu'))

AlexNet.add(layers.MaxPooling2D(pool_size = (3,3), strides = 2, padding = 'valid'))



AlexNet.add(layers.Conv2D(384,(3,3), strides = 1, padding = 'same' ))
AlexNet.add(layers.BatchNormalization())
AlexNet.add(layers.Activation('relu'))

AlexNet.add(layers.Conv2D(384,(3,3), strides = 1, padding = 'same'))
AlexNet.add(layers.BatchNormalization())
AlexNet.add(layers.Activation('relu'))

AlexNet.add(layers.Conv2D(256,(3,3), strides = 1, padding = 'same' ))
AlexNet.add(layers.BatchNormalization())
AlexNet.add(layers.Activation('relu'))
AlexNet.add(layers.MaxPooling2D(pool_size = (3,3), strides = 2, padding = 'valid'))

AlexNet.add(Flatten())

AlexNet.add(layers.Dense(4096))
AlexNet.add(layers.BatchNormalization())
AlexNet.add(layers.Activation('relu'))
#AlexNet.add(Dropout(0.3))

AlexNet.add(layers.Dense(4096))
AlexNet.add(layers.BatchNormalization())
AlexNet.add(layers.Activation('relu'))
#AlexNet.add(Dropout(0.3))

AlexNet.add(layers.Dense(5, activation = 'softmax'))


AlexNet.compile(optimizer= optimizers.SGD(learning_rate=0.01, decay=0.0, momentum=0.0, nesterov=False), loss = 'categorical_crossentropy', metrics = ['accuracy'])
AlexNet.summary()
```
