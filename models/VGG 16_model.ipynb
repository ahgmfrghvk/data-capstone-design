```python
from keras import models
from keras import layers
from keras.layers import Flatten, Dropout
from keras import optimizers, regularizers

VGGNet = models.Sequential()
VGGNet.add(layers.Conv2D(64,(3,3), strides = 1, padding = 'same', input_shape =(224, 224, 3)))
VGGNet.add(layers.BatchNormalization())
VGGNet.add(layers.Activation('relu'))

VGGNet.add(layers.Conv2D(64,(3,3), strides = 1, padding = 'same'))
VGGNet.add(layers.BatchNormalization())
VGGNet.add(layers.Activation('relu'))
VGGNet.add(layers.MaxPooling2D(pool_size = (2,2), strides = 2, padding = 'valid'))


VGGNet.add(layers.Conv2D(128,(3,3), strides = 1, padding = 'same'))
VGGNet.add(layers.BatchNormalization())
VGGNet.add(layers.Activation('relu'))
VGGNet.add(layers.Conv2D(128,(3,3), strides = 1, padding = 'same'))
VGGNet.add(layers.BatchNormalization())
VGGNet.add(layers.Activation('relu'))
VGGNet.add(layers.MaxPooling2D(pool_size = (2,2), strides = 2, padding = 'valid'))


VGGNet.add(layers.Conv2D(256,(3,3), strides = 1, padding = 'same' ))
VGGNet.add(layers.BatchNormalization())
VGGNet.add(layers.Activation('relu'))
VGGNet.add(layers.Conv2D(256,(3,3), strides = 1, padding = 'same'))
VGGNet.add(layers.BatchNormalization())
VGGNet.add(layers.Activation('relu'))
VGGNet.add(layers.Conv2D(256,(3,3), strides = 1, padding = 'same' ))
VGGNet.add(layers.BatchNormalization())
VGGNet.add(layers.Activation('relu'))
VGGNet.add(layers.MaxPooling2D(pool_size = (2,2), strides = 2, padding = 'valid'))


VGGNet.add(layers.Conv2D(512,(3,3), strides = 1, padding = 'same'))
VGGNet.add(layers.BatchNormalization())
VGGNet.add(layers.Activation('relu'))
VGGNet.add(layers.Conv2D(512,(3,3), strides = 1, padding = 'same' ))
VGGNet.add(layers.BatchNormalization())
VGGNet.add(layers.Activation('relu'))
VGGNet.add(layers.Conv2D(512,(3,3), strides = 1, padding = 'same'))
VGGNet.add(layers.BatchNormalization())
VGGNet.add(layers.Activation('relu'))
VGGNet.add(layers.MaxPooling2D(pool_size = (2,2), strides = 2, padding = 'valid'))


VGGNet.add(layers.Conv2D(512,(3,3), strides = 1, padding = 'same' ))
VGGNet.add(layers.BatchNormalization())
VGGNet.add(layers.Activation('relu'))
VGGNet.add(layers.Conv2D(512,(3,3), strides = 1, padding = 'same'))
VGGNet.add(layers.BatchNormalization())
VGGNet.add(layers.Activation('relu'))
VGGNet.add(layers.Conv2D(512,(3,3), strides = 1, padding = 'same'))
VGGNet.add(layers.BatchNormalization())
VGGNet.add(layers.Activation('relu'))
VGGNet.add(layers.MaxPooling2D(pool_size = (2,2), strides = 2, padding = 'valid'))


VGGNet.add(Flatten())

VGGNet.add(layers.Dense(4096))
VGGNet.add(layers.Activation('relu'))
VGGNet.add(Dropout(0.5))

VGGNet.add(layers.Dense(4096))
VGGNet.add(layers.Activation('relu'))
VGGNet.add(Dropout(0.5))

VGGNet.add(layers.Dense(5, activation = 'softmax'))


VGGNet.compile(optimizer= optimizers.SGD(learning_rate=0.01, decay=0.0, momentum=0.0, nesterov=False), loss = 'categorical_crossentropy', metrics = ['accuracy'])
VGGNet.summary()
```
