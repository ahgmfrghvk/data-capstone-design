```python
from keras import models
from keras import layers
from keras.layers import Flatten, Dropout
from keras import optimizers, regularizers

LeNet = models.Sequential()
LeNet.add(layers.Conv2D(6,(5,5), strides = (1,1), padding = 'valid', input_shape =(32, 32, 3), kernel_initializer='he_normal'))
LeNet.add(layers.BatchNormalization())
LeNet.add(layers.Activation('relu'))



LeNet.add(layers.AveragePooling2D(pool_size = (2,2), padding = 'valid'))
LeNet.add(layers.Conv2D(16,(5,5), strides = (1,1), padding = 'valid', kernel_initializer='he_normal'))
LeNet.add(layers.BatchNormalization())
LeNet.add(layers.Activation('relu'))


LeNet.add(layers.AveragePooling2D(pool_size = (2,2), padding = 'valid'))
LeNet.add(layers.Conv2D(120,(5,5), strides = (1,1), padding = 'valid', kernel_initializer='he_normal'))
LeNet.add(layers.BatchNormalization())
LeNet.add(layers.Activation('relu'))


LeNet.add(Flatten())
LeNet.add(layers.Dense(84))
LeNet.add(layers.BatchNormalization())
LeNet.add(layers.Activation('relu'))
#LeNet.add(Dropout(0.3))

LeNet.add(layers.Dense(5, activation = 'softmax'))


LeNet.compile(optimizer= optimizers.SGD(lr=0.005, decay=0.0, momentum=0.0, nesterov=False), loss = 'categorical_crossentropy', metrics = ['accuracy'])
LeNet.summary()
```
